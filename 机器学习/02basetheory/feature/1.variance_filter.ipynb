```python
import numpy as np
```

```python
# 构造特征
a = np.random.randn(100)
print(a.var())
```

```python
b = np.random.randn(100) * 0.1
b = np.random.normal(5, 0.1, 100)
print(b.var())
```

```python
# 构造特征数据
X = np.vstack((a,b)).T
print(X.shape)
print(X)
```

```python
# 低方差过滤
from sklearn.feature_selection import VarianceThreshold
```

```python
vt = VarianceThreshold(threshold=0.01)
X_filted = vt.fit_transform(X)
print(X_filted.shape)
```
