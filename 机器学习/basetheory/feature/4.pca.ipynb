```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.proj3d import proj_transform
from sklearn.decomposition import PCA
```

```python
X = np.random.randn(1000,3)
print(X.shape)
```

```python
# 使用 PCA 将三维数据降为两维
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print(X_pca.shape)
```

```python
# 可视化
# 转换前的3维数据可视化
fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(121,projection='3d')
ax1.scatter(X[:,0],X[:,1],X[:,2],c='g')
ax1.set_title('before PCA(3D)')
ax1.set_xlabel('Feature1')
ax1.set_label('Feature2')
ax1.set_zlabel('Feature3')

# 降维后的数据可视化
ax2 = fig.add_subplot(122)
ax2.scatter(X_pca[:,0],X_pca[:,1],c='g')
ax2.set_title('after PCA(3D)')
ax2.set_xlabel('Principal Component1')
ax2.set_label('Principal Component2')
plt.show()
```

```python
# 手动构建线性相关的三组特征数据
n = 1000
pc1 = np.random.normal(0,1,n)
pc2 = np.random.normal(0,0.1,n)
# 定义不重要的第三主成分（噪声）
noise = np.random.normal(0,0.05,n)
# 构建特征矩阵
X = np.vstack((pc1 + pc2,pc1 - pc2,pc2 + noise)).T
print(X.shape)
```

```python
# 使用 PCA 将三维数据降为两维
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print(X_pca.shape)
```

```python
# 可视化
# 转换前的3维数据可视化
fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(121,projection='3d')
ax1.scatter(X[:,0],X[:,1],X[:,2],c='g')
ax1.set_title('before PCA(3D)')
ax1.set_xlabel('Feature1')
ax1.set_ylabel('Feature2')
ax1.set_zlabel('Feature3')

# 降维后的数据可视化
ax2 = fig.add_subplot(122)
ax2.scatter(X_pca[:,0],X_pca[:,1],c='g')
ax2.set_title('after PCA(3D)')
ax2.set_xlabel('Principal Component1')
ax2.set_label('Principal Component2')
plt.show()
```
