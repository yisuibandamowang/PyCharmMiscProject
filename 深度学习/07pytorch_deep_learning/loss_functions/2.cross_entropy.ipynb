```python
import torch
import torch.nn as nn
```

```python
# 1. 目标值为顺序编码的标签
input = torch.randn(6, 8)
print(input)
# pred = torch.softmax(input, dim=1)
# print(pred)
```

```python
# 目标值，6个数据的类别标签（0~7）
target = torch.tensor([1, 0, 3, 7, 5, 2])
print(target)
```

```python
# 定义损失函数
loss = nn.CrossEntropyLoss()
print(loss(input, target))
```

```python
# 2. 目标值为一组概率（独热编码）
input = torch.randn(6, 8)
print(input)
```

```python
target = torch.randn(6, 8).softmax(dim=1)
print(target)
```

```python
loss = nn.CrossEntropyLoss()
print(loss(input, target))
```
