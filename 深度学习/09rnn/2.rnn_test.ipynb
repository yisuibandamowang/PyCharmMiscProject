```python
import torch
import torch.nn as nn
```

```python
# 定义RNN层
rnn = nn.RNN(input_size=8, hidden_size=16, num_layers=2, batch_first=True)
```

```python
# 定义输入数据和初始隐状态
input = torch.randn(3, 3, 8)
hx = torch.randn(2, 3, 16)
print(input)
print(hx)
```

```python
# 前向传播，得到输出和隐状态
output, hn = rnn(input, hx)
print(output)
print(hn)
print(output.shape)
print(hn.shape)
```
