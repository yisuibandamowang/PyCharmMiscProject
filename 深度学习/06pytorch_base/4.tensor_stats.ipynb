```python
import torch
```

```python
tensor1 = torch.randint(1, 10, (3, 2, 4)).float()
print(tensor1)
```

```python
# sum 求和
print(tensor1.sum())
```

```python
print(tensor1.sum(dim=2))
```

```python
# mean 求均值
print(tensor1.mean())
```

```python
print(tensor1.mean(dim=2))
```

```python
# std 求标准方差
print(tensor1.std())
```

```python
# max、min求最大最小值
print(tensor1.max())
```

```python
print(tensor1.min(dim=2))
```

```python
print(tensor1.argmax())
```

```python
# 去重
print(tensor1.unique())
```

```python
# 排序
print(tensor1.sort())
```
