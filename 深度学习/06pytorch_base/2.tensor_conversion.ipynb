# 张量类型转换


```python
tensor1 = torch.tensor([1,2,3])
print(tensor1.dtype)
```

```python
tensor1 = tensor1.type(torch.float64)
print(tensor1.dtype)
```

```python
tensor1 = tensor1.half()
print(tensor1.dtype)
```

```python
print(tensor1)
```

```python
tensor1 = tensor1.to(torch.complex64)
print(tensor1, tensor1.dtype)
```
