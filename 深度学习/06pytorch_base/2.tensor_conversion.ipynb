# 张量类型转换


```python
tensor1 = torch.tensor([1,2,3])
print(tensor1.dtype)
```

```python
tensor1 = tensor1.type(torch.float64)
print(tensor1.dtype)
```

```python
tensor1 = tensor1.half()
print(tensor1.dtype)
```

```python
print(tensor1)
```

```python
tensor1 = tensor1.to(torch.complex64)
print(tensor1, tensor1.dtype)
```

# Tensor 和 ndarray 的转换


```python
np.set_printoptions(precision=6)
torch.set_printoptions(precision=6)

# 1. tensor → ndarray
tensor1 = torch.rand(3, 2)
print(tensor1)
```

```python
ndarray1 = tensor1.numpy()
print(ndarray1)
```

```python
tensor1[:, 0] = 10
print(tensor1)
print(ndarray1)
```

```python
# 使用copy方法避免共享内存
ndarray2 = tensor1.numpy().copy()
print(ndarray2)
```

```python
tensor1[:, 0] = 5
print(tensor1)
print(ndarray1)
print(ndarray2)
```

```python
# 2. ndarray → tensor
ndarray1 = np.random.randn(3, 2)
tensor1 = torch.from_numpy(ndarray1)
print(tensor1)
print(ndarray1)
```

```python
ndarray1[:, 0] = 5
print(tensor1)
print(ndarray1)
```

```python
# 使用copy方法避免内存共享
tensor2 = torch.from_numpy(ndarray1.copy())
ndarray1[:, 0] = 7
print(ndarray1)
print(tensor1)
print(tensor2)
```

```python
# 使用tensor方法直接转换成Tensor
ndarray1 = np.random.randn(3, 2)
tensor1 = torch.tensor(ndarray1)
print(tensor1)
print(ndarray1)
```

```python
ndarray1[:, 0] = 5
print(tensor1)
print(ndarray1)
```
