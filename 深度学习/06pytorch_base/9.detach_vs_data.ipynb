```python
import torch
```

```python
# 用两组数据进行对比
x1 = torch.tensor([1.0, 2, 3], requires_grad=True)
x2 = torch.tensor([1.0, 2, 3], requires_grad=True)
print(x1)
print(x2)
```

```python
y1 = x1.sigmoid()
y2 = x2.sigmoid()
print(y1)
print(y2)
```

```python
z1 = y1.data
z2 = y2.detach()
print(z1)
print(z2)
```

```python
print(z1.requires_grad)
print(z2.requires_grad)
```

```python
z1.zero_()
z2.zero_()
print(y1)
print(y2)
```

```python
y1.sum().backward()
print(x1.grad)
```

```python
y2.sum().backward()
print(x2.grad)
```
